```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sbn
import json
import os
import sys
import pandas as pd
import pickle

root_path = os.environ['FCLA_ROOT']
data_path = os.environ['DATA_PATH']
sys.path.append(os.path.join(root_path,'scripts'))

from mergeResults import *
```

```python
results = load_results(os.path.join(data_path, "geometric", '30sourcesVarySize', "solutions", "gurobi"), 
                       os.path.join(data_path,'geometric','30sourcesVarySize'))
```

```python
for col in results.columns:
    print('-----')
    print(results[col].value_counts())
```

```python
results[['id','vcount','objective','source_num','number of facilities']]
```

```python
results_fcla = load_results(os.path.join(data_path, "geometric", '30sourcesVarySize', "solutions", "fcla"), 
                            os.path.join(data_path,'geometric', '30sourcesVarySize'))
```

```python
results_fcla['algorithm'] = 'wma'
results['algorithm'] = 'gurobi'
```

```python
all_results = pd.concat([results, results_fcla], axis=0)
```

```python
all_results
```

```python
# plt.gca().set_xscale("log", nonposx='clip')
fig, ax = plt.subplots(figsize=(8, 10), dpi=80)
plt.subplot(2, 1, 1)
plt.gca().set_yscale("log", nonposy='clip')

results_fac = results[results['vcount'] == 1000].sort_values(by='number of facilities')
results_fac_fcla = results_fcla[results_fcla['vcount'] == 1000].sort_values(by='number of facilities')
plt.plot(results_fac['number of facilities'],results_fac['runtime'],label="gurobi")
plt.plot(results_fac_fcla['number of facilities'],results_fac_fcla['runtime'],label='wma')
plt.xlabel("number of facilities")
plt.ylabel("runtime, log seconds")
plt.title("30 source nodes, network 1000 nodes; capacity = 5")
plt.legend()


plt.subplot(2, 1, 2)
results_fac = results[results['vcount'] == 1000].sort_values(by='number of facilities')
results_fac_fcla = results_fcla[results_fcla['vcount'] == 1000].sort_values(by='number of facilities')
plt.plot(results_fac['number of facilities'],results_fac['objective'],label="gurobi")
plt.plot(results_fac_fcla['number of facilities'],results_fac_fcla['objective'],label='wma')
plt.xlabel("number of facilities")
plt.ylabel("objective")
plt.title("30 source nodes, network 1000 nodes; capacity = 5")
plt.legend()
```

```python
# plt.gca().set_xscale("log", nonposx='clip')
fig, ax = plt.subplots(figsize=(8, 10), dpi=80)
plt.subplot(2, 1, 1)
plt.gca().set_yscale("log", nonposy='clip')

results_fac = results[results['number of facilities'] == 8].sort_values(by='vcount')
results_fac_fcla = results_fcla[results_fcla['number of facilities'] == 8].sort_values(by='vcount')
plt.plot(results_fac['vcount'],results_fac['runtime'],label="gurobi")
plt.plot(results_fac_fcla['vcount'],results_fac_fcla['runtime'],label='wma')
plt.xlabel("network size")
plt.ylabel("runtime, log seconds")
plt.title("30 source nodes, 8 facilities; capacity = 5")
plt.legend()


plt.subplot(2, 1, 2)
results_fac = results[results['number of facilities'] == 8].sort_values(by='vcount')
results_fac_fcla = results_fcla[results_fcla['number of facilities'] == 8].sort_values(by='vcount')
plt.plot(results_fac['vcount'],results_fac['objective'],label="gurobi")
plt.plot(results_fac_fcla['vcount'],results_fac_fcla['objective'],label='wma')
plt.xlabel("network size")
plt.ylabel("objective")
plt.title("30 source nodes, 8 facilities; capacity = 5")
plt.legend()
```
